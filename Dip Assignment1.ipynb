```python
def gamma_transformation(img, gamma):
    normalize_img = img/255.0
    gamma_corrected= np.power(normalize_img, gamma)
    return gamma_corrected

gamma_img1= gamma_transformation(img,1)
gamma_img2= gamma_transformation(img,3)
gamma_img3= gamma_transformation(img,0.4)
plt.figure(figsize=(15, 8))

    # Original image
plt.subplot(1, 4, 1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

    # Gamma = 1
plt.subplot(1, 4, 2)
plt.imshow(gamma_img1)
plt.title('Gamma = 1')
plt.axis('off')

    # Gamma = 3
plt.subplot(1, 4, 3)
plt.imshow(gamma_img2)
plt.title('Gamma = 3')
plt.axis('off')

    # Gamma = 0.4
plt.subplot(1, 4, 4)
plt.imshow(gamma_img3)
plt.title('Gamma = 0.4')
plt.axis('off')
plt.show()
```
